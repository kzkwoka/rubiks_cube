```python
%matplotlib notebook
import numpy as np
import cv2
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


images_path = './images/'
# scaling factor from opencv to (360deg,100%,100%) H*2,S/255*100,V/255*100
```

```python
blues = []
for i in range(1,7):
    name = f"{i}.jpg"
    image = cv2.imread(images_path + name )[230:300,250:310]
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    mean = cv2.mean(hsv)
    blues.append([mean[0]*2,mean[1]/255*100,mean[2]/255*100])
blues = np.array(blues)
print(blues)
```

```python
reds = []
for i in range(1,7):
    name = f"{i}.jpg"
    image = cv2.imread(images_path + name )[230:300,170:220]
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    mean = cv2.mean(hsv)
    reds.append([mean[0]*2,mean[1]/255*100,mean[2]/255*100])
reds = np.array(reds) ## weird pink results
print(reds)
```

```python
whites = []
for i in range(1,7):
    name = f"{i}.jpg"
    image = cv2.imread(images_path + name )[240:290,80:140]
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    mean = cv2.mean(hsv)
    whites.append([mean[0]*2,mean[1]/255*100,mean[2]/255*100])
whites = np.array(whites)
print(whites)
```

```python
greens = []
for i in range(1,7):
    name = f"{i}.jpg"
    image = cv2.imread(images_path + name )[150:200,170:220]
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    mean = cv2.mean(hsv)
    greens.append([mean[0]*2,mean[1]/255*100,mean[2]/255*100])
greens = np.array(greens)
print(greens)
```

```python
oranges = []
for i in range(1,7):
    name = f"{i}.jpg"
    image = cv2.imread(images_path + name )[50:110,170:220]
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    mean = cv2.mean(hsv)
    oranges.append([mean[0]*2,mean[1]/255*100,mean[2]/255*100])
oranges = np.array(oranges)
print(oranges)
```

```python
yellows = []
for i in range(1,7):
    name = f"{i}.jpg"
    image = cv2.imread(images_path + name )[50:110,250:310]
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    mean = cv2.mean(hsv)
    yellows.append([mean[0]*2,mean[1]/255*100,mean[2]/255*100])
yellows = np.array(yellows)
print(yellows)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(yellows[:,0],yellows[:,2],yellows[:,1], c='yellow', marker='o')
ax.scatter(oranges[:,0],oranges[:,2],oranges[:,1], c='orange', marker='o')
ax.scatter(reds[:,0],reds[:,2],reds[:,1], c='r', marker='o')
ax.scatter(greens[:,0],greens[:,2],greens[:,1], c='g', marker='o')
ax.scatter(blues[:,0],blues[:,2],blues[:,1], c='b', marker='o')
ax.scatter(whites[:,0],whites[:,2],whites[:,1], c='black', marker='o')
ax.set_xlabel('H')
ax.set_ylabel('V')
ax.set_zlabel('S')
plt.show()
```
